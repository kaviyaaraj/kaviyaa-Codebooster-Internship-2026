```python
df['total_score']=df['math_score']+df['science_score']+df['english_score']+df['programming_score']
```

```python
top_student=df.loc[df['total_score'].idxmax()]
```

```python
print("Top Student:")
print(f"Name:{top_student['name']}")
print(f"Department:{top_student['department']}")
print(f"Total Score:{top_student['total_score']}")
print(f"Attendance:{top_student['attendance_percentage']}")
```

```python
dept_avg=df.groupby('department')['total_score'].mean().round(2)
```

```python
print("Average Total Score by Department:")
print(dept_avg)
```

```python
df_sorted=df.sort_values(by='total_score',ascending=False)
```

```python
print("Top 10 Students by Total Score:")
print(df_sorted[['name','department','total_score','attendance_percentage']].head(10))
```

```python
print("\nDATASET OVERVIEW")
print("-" * 30)
print("Total Number of Students:", len(df))
print("Total Number of Columns:", len(df.columns))
print("Number of Unique Departments:", df['department'].nunique())

```

```python
print("\nDEPARTMENT-WISE ANALYSIS")
print("-" * 30)
print(df['department'].value_counts())
```

```python
print("\nSCORE ANALYSIS")
print("-" * 30)
print("Highest Math Score:", df['math_score'].max())
print("Lowest Math Score:", df['math_score'].min())
print("Average Math Score:", df['math_score'].mean())
print("Average Programming Score:", df['programming_score'].mean())
print("Average Attendance Percentage:", df['attendance_percentage'].mean())
```

```python
print("\nGENDER-WISE ANALYSIS")
print("-" * 30)
gender_avg = df.groupby('gender')['total_score'].mean()
print(gender_avg)
```

```python
print("\nTOP 5 STUDENTS")
print("-" * 30)

top_students = df.sort_values(by='total_score', ascending=False).head(5)

for index, student in top_students.iterrows():
    print(f"Name: {student['name']}")
    print(f"Department: {student['department']}")
    print(f"Total Score: {student['total_score']}")
    print("-" * 20)
```

```python
print("\nATTENDANCE ANALYSIS")
print("-" * 30)

low_attendance = df[df['attendance_percentage'] < 75]

if low_attendance.empty:
    print("All students have attendance above 75%")
else:
    for index, student in low_attendance.iterrows():
        print(f"Name: {student['name']}")
        print(f"Department: {student['department']}")
        print(f"Attendance Percentage: {student['attendance_percentage']}%")
        print("-" * 20)
```
